# Aggregation and Visualization

```python
df4= df.groupby('Continent')[['1970 Population','1980 Population','1990 Population','2000 Population','2010 Population','2015 Population','2020 Population','2022 Population']].mean()
df4
```

```python
# Transpose for Visualization

df3= df4.T
df3
```
